# Exploring with Visuals

Use histograms and scatterplots to explore 'clean_08.csv' and 'clean_18.csv' in the Jupyter notebook. Then, answer the quiz questions below the notebook.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
print('Import Libraries Complete')
```

```python
# dataset url
url_08 = ('https://raw.githubusercontent.com/bentegviz/udacity_intro_to_data_analysis/main/Case%20Study%202/data/processed/clean_08.csv')
url_18 = ('https://raw.githubusercontent.com/bentegviz/udacity_intro_to_data_analysis/main/Case%20Study%202/data/processed/clean_18.csv')
```

```python
# csv to dataset
df_08 = pd.read_csv(url_08)
df_18 = pd.read_csv(url_18)
print('Load CSV Complete')
```

```python
df_08.head()
```

```python
df_18.head()
```

Histograms

```python
# 2008 Dataset Histograms
sns.histplot(data=df_08)
```

```python
# 2008 Greenhouse Gas Scores
sns.histplot(data=df_08['greenhouse_gas_score'], bins=10)
```

```python
# 2018 Greenhouse Gas Scores
sns.histplot(data=df_18['greenhouse_gas_score'], bins=10)
```

```python
# 2008 Greenhouse Gas Scores
sns.histplot(data=df_08['cmb_mpg'], bins=10)
```

```python
# 2018 Greenhouse Gas Scores
sns.histplot(data=df_18['cmb_mpg'], bins=20)
```

Correlations

```python
# Compute the Correlation Matrix
corr_08 = df_08.corr()
corr_18 = df_18.corr()
```

```python
# Correlation between Displacement and Combined MPG
sns.heatmap(corr_08, center=0, square=True, vmax=1.0, vmin=-1.0)
```

```python
# Correlation between Displacement and Combined MPG
sns.heatmap(corr_18, square=True, vmax=1.0, vmin=-1.0)
```

 Scatterplots

```python
# 2008 Dataset Histograms
sns.scatterplot(data=df_08)
```

```python
# 2008 Dataset Histograms
sns.scatterplot(data=df_08['air_pollution_score'])
```

```python
# 2018 Dataset Histograms
sns.scatterplot(data=df_18)
```

```python
# 2018 Dataset Histograms
sns.scatterplot(data=df_18['air_pollution_score'])
```
